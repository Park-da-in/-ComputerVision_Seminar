```python
import glob
import cv2 #computer vision
import matplotlib.pyplot as plt # 시각화
import numpy as np
from google.colab.patches import cv2_imshow # cv2.imshow(img)가 안되므로 
from PIL import Image as PILImage # PIL = Pithon Image Library

image_1 = (cv2.imread(dolphin_img_list[-1])) # -> (H, W, C) : C축이 RGB가 아니라 BGR로 읽힌다.
print(image_1.shape)
cv2_imshow(image_1) # size = (259, 194, 3)
```

```python
(b, g, r) = image_1[0,0]
print("Pixel at (0, 0) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
cv2_imshow(image_1)

(b, g, r) = image_1[129, 97]
print("Pixel at (129, 97) - Red: {}, Green: {}, Blue: {}".format(r, g, b))
cv2_imshow(image_1)

image_1[129, 97] = (0, 0, 255)
(b, g, r) = image_1[129, 97]
print("Changed Pixel at (129, 97) - Red: {}, Green: {}, Blue: {}".format(r, g, b))

cv2_imshow(image_1)
```

```python
(h, w) = image_1.shape[:2]

(cX, cY) = (w//2, h//2)
tl = image_1[0:cY, 0:cX]
cv2_imshow(tl)
```

```python
image_1[0:cY, 0:cX] = (255, 255, 0)
 
# Show our updated image
cv2_imshow(image_1)
```
